```python
# this sim was not filtered for detection, so just do a rough cut at the single visit depth
single_visit_depth = {'u':23.9,'g':25.0,'r':24.7,'i':24.0,'z':23.3,'y':22.1}
```

```python
def get_diaSources_from_ssObjectId(ssObjectId, connection):
    df =  pd.read_sql_query(f'select ssObjectId, diaSourceId, ccdVisitId, ra, decl, mag, midPointTai, filter from diaSource where ssObjectId = {ssObjectId} ORDER BY midPointTai', connection)
    df = df.astype({'ssObjectId':'int64', 'diaSourceId':'int64', 'ccdVisitId':'int64', 
                    'ra':'float32', 'decl':'float32', 'mag':'float16', 'midPointTai':'float32'})
    df = df.rename({'filter':'filterName'},axis='columns')
    keep = pd.Series(False,index=df.index)
    for band, depth in single_visit_depth.items():
        keep |= (df['filterName'] == band) & (df['mag'] <= depth)
    
    return df.loc[keep].to_dict('records')
```

```python
def patchDiaSource(ds):
    #diaSource fixes
    ds['programId'] = 0
    ds['flags'] = 0

    
    required = ['apFlux', 'apFluxErr', 'psFlux', 'psFluxErr', 'x', 'y','snr']
    for require in required:
        if require not in ds:
            ds[require] = np.nan
    
    # this is to give us *some* information while we wait for the detection SNR
    #ds['snr'] = ds['apFlux']/ds['apFluxErr']

    return ds

def patchSSObject(ss, numObs):
    ss['flags'] = 0
    
    # replace the number of obs in the database (which did not consider detectability) with the number brighter than
    # the single visit depth
    ss['numObs'] = np.int16(numObs)

    return ss

def make_alert_json(ssObject, connection):
    alert = {}
    alert['alertId'] = 0
    
    ssObject_dict = ssObject.to_dict(orient='records')[0]
    
    diaSources = get_diaSources_from_ssObjectId(ssObject_dict['ssObjectId'], connection)
    if len(diaSources) == 0:
        return None
    
    # latest diaSource is last
    alert['diaSource'] = patchDiaSource(diaSources[-1])

    alert['prvDiaSources'] = [patchDiaSource(ds) for ds in diaSources[:-1]]
    
    alert['prvDiaForcedSources'] = None
    alert['prvDiaNondetectionLimits'] = None
    
    alert['diaObject'] = None
    
    alert['ssObject'] = patchSSObject(ssObject_dict, len(diaSources))
    
    alert['cutoutDifference'] = None
    alert['cutoutTemplate'] = None
    
    return alert
```

```python
with open(f'simulated_sso_alerts_filtered.avro','a+b') as f:
    alerts = []
    with ProgressBar(len(ssobjects), ipython_widget=True) as bar:
        for idx, _ in ssobjects.iterrows():
            # don't use the row in the iterable directly because numpy coerces everything to floats
            alert_json = make_alert_json(ssobjects.loc[idx:idx+1], connection)
            if alert_json is not None:
                alerts.append(alert_json)
            bar.update()
            if len(alerts) == 1000:
                schema.store_alerts(f,alerts)
                alerts = []
        schema.store_alerts(f,alerts)
```
